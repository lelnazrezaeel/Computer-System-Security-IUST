<h1>
<font color="#faf200">Watermarking</font>
</h1>

```python
import random
from PIL import Image, ImageDraw, ImageFont
import cv2 as cv
import numpy as np
```

<h2>
<font color="#ad41fa">Opening image & creating new text layer</font>
</h2>

```python
img = Image.open('2.jpg').convert("RGBA")
txt = Image.new('RGBA', img.size, (255,255,255,0))
```

<h2>
<font color="#ad41fa">Diplay the original image</font>
</h2>

```python
original = Image.open('2.jpg')
original = original.resize((256,256))
original.show()
```

<h2>
<font color="#ad41fa">Creating text</font>
</h2>

```python
text = "Security_IUST_1402"
font = ImageFont.truetype("arial.ttf", 27)
```

<h2>
<font color="#ad41fa">Creating draw object</font>
</h2>

```python
draw = ImageDraw.Draw(txt)
```

<h2>
<font color="#ad41fa">Positioning of text</font>
</h2>

```python
width, height = img.size 
```

<h2>
<font color="#ad41fa">Loop for multiple watermarks</font>
</h2>

```python
y=200
for i in range(7):
    x=random.randint(0, width-300)
    y+=random.randrange(0,int(height/8), 19)+random.randint(0,100)
    draw.text((x,y), text, fill=(255,255,255, 75), font=font)
```

<h2>
<font color="#ad41fa">Combining both layers and saving new image</font>
</h2>

```python
watermarked = Image.alpha_composite(img, txt)
watermarked.save('watermarked.png')
```

<h2>
<font color="#ad41fa">Diplay the watermarked image</font>
</h2>

```python
watermarked = Image.open('watermarked.png')
watermarked = watermarked.resize((256,256))
watermarked.show()
```

<h2>
<font color="#ad41fa">Opening watermarked image</font>
</h2>

```python
img = cv.imread('watermarked.png')
```

<font color="#ad41fa">Rotate the image</font>
</h2>

```python
rotated = cv.rotate(img, cv.ROTATE_90_CLOCKWISE)
cv.imwrite('rotated.png', rotated)
```

<font color="#ad41fa">Diplay the rotated image</font>
</h2>

```python
rotated = Image.open('rotated.png')
rotated = rotated.resize((256,256))
rotated.show()
```

<font color="#ad41fa">Resize the image</font>

<h2>
<font color="#ad41fa">Add noise to the image</font>
</h2>

```python
gauss_noise=np.zeros((1250,1920),dtype=np.uint8)
cv.randn(gauss_noise,128,20)
gauss_noise=(gauss_noise*0.5).astype(np.uint8)
imgGray = cv.imread('watermarked.png', 0)
gn_img = cv.add(imgGray, gauss_noise)
cv.imwrite('noisy.png', gn_img)
```

<h2>
<font color="#ad41fa">Diplay the noisy image</font>
</h2>

```python
noisy = Image.open('noisy.png')
noisy = noisy.resize((256,256))
noisy.show()
```
